#### Define paths

````python
main_path = Path(r"X:\Hussein_SOD1") # Main directory
target_rel = Path("preprocessed") / "kilosort4" # Target relative path (platform independent)
oebin_mainPath = Path(r'I:\Neuropixels_Backup\Hussein')
ks4_paths = [
    p for p in main_path.rglob("kilosort4")
    if p.is_dir() and p.parent.name == "preprocessed"
]
oebin_paths = [p for p in oebin_mainPath.rglob("structure.oebin") if p.is_file()]
print('all oebin paths:')
pprint(oebin_paths)
````

````python

# ks_dir = os.path.normpath(r"X:\Hussein_SOD1\20250702_mCherry_SOD1_N01\2025-07-02_14-42-04\preprocessed\kilosort4")
for ks_dir in ks4_paths:
    ks_dir = os.path.join(ks_dir,'sorter_output') # sorter_output if Ks was run from spikeinterface, else ks_dir is just the kilosort4 folder
    save_path = Path(ks_dir) / "bombcell" # Set bombcell's output directory
    search_path = Path(ks_dir).parent.parent.parent
    # grab the name of last folder in path 'search_path'
    rec_name = search_path.name
    # Step 2: filter by paths that contain the common folder in the expected position
    matching_oebin = []
    for f in oebin_paths:
        # Check if the common folder name exists in the parts of the path
        if rec_name in f.parts:
            matching_oebin.append(f)
            break # stop after finding the first match (assuming one match per recording)
    ## Provide raw and meta files
    ## Leave 'None' if no raw data. Ideally, your raw data is common-average-referenced and
    # the channels are temporally aligned to each other (this can be done with CatGT)
    # look for the file inside the ks_dir or provide full path here:

    params_path = os.path.join(ks_dir, "params.py")
    
    if os.path.exists(params_path):
        print(params_path)
    else:
        print("params.py not found")

    spec = importlib.util.spec_from_file_location("params", params_path)
    params = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(params)
    raw_file_path = params.dat_path[0]  # assuming single raw file
    # raw_file_path = os.path.normpath(r'X:\NP1_chronic_tACS_NIH\preprocessed\traces_cached_seg0.raw') #None#"/home/julie/Dropbox/Example datatsets/JF093_2023-03-09_site1/site1/2023-03-09_JF093_g0_t0_bc_decompressed.imec0.ap.bin" # ks_dir
    meta_file_path = matching_oebin[0] if matching_oebin else None #None#"/home/julie/Dropbox/Example datatsets/JF093_2023-03-09_site1/site1/2023-03-09_JF093_g0_t0_bc_decompressed.imec0.ap.bin"None#"/home/julie/Dropbox/Example datatsets/JF093_2023-03-09_site1/site1/2023-03-09_JF093_g0_t0.imec0.ap.meta"
    if meta_file_path is None:
        print("No matching oebin file found for recording:", rec_name)
        continue
    ## Get default parameters - we will see later in the notebook how to assess and fine-tune these
    param = bc.get_default_parameters(ks_dir, 
                                    raw_file=raw_file_path,
                                    meta_file=meta_file_path,
                                    kilosort_version=4)

    print("Bombcell parameters:")
    pprint(param)
    # you might to change:

    # 1. classification thresholds like: 
    param["maxRPVviolations"] = 0.2 # default is 0.1, but it depends on rec, keep high initially to see how the quality metrics look and then decrease if needed.
    param["minSNR"] = 1 # default is 5, but it depends on rec, keep low initially to see how the quality metrics look and then increase if needed.     
    param["minAmplitude"] = 0.5 # in microvolts, default is 10, but it depends on rec, keep low initially to see how the quality metrics look and then increase if needed.
    param["maxPeak1ToPeak2Ratio_nonSomatic"] = 4
    param["flatnessThreshold"] = 0.5
    param["maxWvBaselineFraction"] = 0.6 # default is 0.3, but too low tends to exclude small amplitude units 
    #  2. or which quality metrics are computed (by default these are not): 
    param["computeDistanceMetrics"] = 1
    param["computeDrift"] = 1
    param["splitGoodAndMua_NonSomatic"] = 0

    #  3. how quality metricsa are calculated:
    # a. how refractory period window is defined
    param["tauR_valuesMin"]= 0.5 / 1000  # minumum refractory period time (s), usually 0.002 s
    param["tauR_valuesMax"]= 5 / 1000  # maximum refractory period time (s)
    param["tauR_valuesStep"]= 0.5 / 1000  # if tauR_valuesMin and tauR_valuesMax are different, bombcell 
    # will calculate refractory period violations from param["tauR_valuesMin"] to param["tauR_valuesMax"] param["tauR_valuesStep"] 
    # bins and determine the option window for each unit before calculating the violations. 
            # tauR_valuesStep
    #  b. or whether the recording is split into time chunks to detemrine "good" time chunks: 
    # param["computeTimeChunks"] = 0
    # full list in the wiki or in the bc.get_default_parameters function
    try:
        (
            quality_metrics,
            param,
            unit_type,
            unit_type_string,
        ) = bc.run_bombcell(
            ks_dir, save_path, param
        )
        print(f"Finished processing {ks_dir} folder, results saved in {save_path}")
    except Exception as e:
        print(f"Error processing {ks_dir} folder, cannot write in {save_path}: {e}")
        print(e)
        continue
    # Use the output summary plots (below) to see if the 
    # quality metric thresholds seem roughly OK for your 
    # data (i.e. there isn't one threshold removing all 
    # units or a threshold may below that removes none)
    # more details on these output plots in the wiki:
    # https://github.com/Julie-Fabre/bombcell/wiki/Summary-output-plots
````
